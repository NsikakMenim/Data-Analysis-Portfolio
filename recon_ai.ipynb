```python
import pandas as pd
import numpy as np

def load_and_clean_data(filepath):
    # Load the CSV file into a DataFrame
    try:
        df = pd.read_csv(filepath)
    except FileNotFoundError:
        print("File not found. Please check the file path.")
        return None

    # Standardize column names (trim whitespace)
    df.columns = df.columns.str.strip()

    # Drop the 'Campaigns' column if it exists
    if 'Campaigns' in df.columns:
        df = df.drop('Campaigns', axis=1)

    # Fill missing numeric volume values with 0
    volume_cols = ['Min Monthly Volume', 'Max Monthly Volume', 'Monthly Volume']
    for col in volume_cols:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors='coerce').fillna(0)

    return df

def infer_search_intent(keyword):
    # Convert keyword to lowercase for case-insensitive matching
    kw = str(keyword).lower()
    # Heuristic: if keyword contains question words, assume Informational
    if any(q in kw for q in ["what", "how", "why"]):
        return "Informational"
    # Heuristic: if keyword contains words like best, compare, top, buy, software, assume Transactional
    elif any(t in kw for t in ["best", "compare", "top", "buy", "software"]):
        return "Transactional"
    # Default to Informational
    return "Informational"

def fill_search_intent(df):
    # Fill missing or empty values in 'Search Intent' using our heuristic
    df['Search Intent'] = df['Search Intent'].fillna("").str.strip()
    df['Search Intent'] = df.apply(
        lambda row: infer_search_intent(row['Keyword']) if row['Search Intent'] == '' else row['Search Intent'],
        axis=1
    )
    return df

def parse_dates_and_fill(df):
    # Parse the 'Last Updated' column as datetime
    df['Last Updated'] = pd.to_datetime(df['Last Updated'], errors='coerce')
    # Fill missing dates with a default date
    default_date = pd.to_datetime('2024-12-31')
    df['Last Updated'] = df['Last Updated'].fillna(default_date)
    return df

def map_difficulty(df):
    # Ensure the 'Difficulty' column is numeric and fill missing with the median
    df['Difficulty'] = pd.to_numeric(df['Difficulty'], errors='coerce')
    median_difficulty = df['Difficulty'].median()
    df['Difficulty'] = df['Difficulty'].fillna(median_difficulty)

    # Map numeric difficulty to qualitative rating using defined thresholds:
    # 1–39: Low, 40–60: Moderate, 61–100: High
    def difficulty_rating(score):
        if pd.isna(score):
            return "Unknown"
        elif 1 <= score <= 39:
            return "Low"
        elif 40 <= score <= 60:
            return "Moderate"
        elif 61 <= score <= 100:
            return "High"
        else:
            return "Unknown"

    df['Difficulty Rating'] = df['Difficulty'].apply(difficulty_rating)
    return df

def remove_duplicates(df):
    # Remove duplicate rows
    return df.drop_duplicates()

def main():
    filepath = "reconcile_ai_data.csv"

    # Load and initial clean-up
    df = load_and_clean_data(filepath)
    if df is None:
        return

    # Fill missing 'Search Intent' values
    df = fill_search_intent(df)

    # Parse dates in 'Last Updated' and fill missing values
    df = parse_dates_and_fill(df)

    # Map numeric Difficulty to a qualitative rating and fill missing values
    df = map_difficulty(df)

    # Remove duplicates
    df = remove_duplicates(df)

    # Display the first few rows of the cleaned DataFrame
    print("Optimized and Cleaned DataFrame:")
    print(df.head())

    # Save the cleaned DataFrame to a new CSV file
    df.to_csv("reconcile_ai_data_cleaned_optimized.csv", index=False)

if __name__ == "__main__":
    main()
```
