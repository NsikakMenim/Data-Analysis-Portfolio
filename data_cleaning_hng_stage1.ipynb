```python
def generate_short_title(title):
    """
    Generates concise product titles (30-50 characters) by:
    - Returning titles less than 7 words.
    - Removing redundant phrases(set of|pack of|combo of|includes|features|for|with|and)
    - Preserving key product details(colour, quantity, size, demographics and clothing term)
    - Maintaining essential keywords and phrases.
    - Truncating to a limit of 50 characters
    -
    """

    # Immediate return for short titles (<7 words)
    words = title.split()
    if len(words) < 7:
        return title

    # Predefined lists
    REDUNDANT_PHRASES = r'\b(set of|pack of|combo of|includes|features|for|with|and)\b'
    DEMOGRAPHICS = {'men', 'women', 'girls', 'boys', 'male', 'female', 'children'}
    CLOTHING_TERMS = {'legging', 'shirt', 'trouser', 'gown', 'dress', 'jacket',
                     'jeans', 'skirt', 'shorts', 'sneaker', 'loafer', 'curtain',
                     'pyjama', 'ankle', 'set', 'combo', 'shoe', 'slipper'}

    # Step 1: Remove redundant phrases
    clean_title = re.sub(REDUNDANT_PHRASES, '', title, flags=re.IGNORECASE)

    # Step 2: Split into meaningful segments
    segments = re.split(r'[|_,:&]', clean_title)

    # Step 3: Extract key components
    components = {
        'color': re.search(r'\b(black|navy|blue|red|green|white)\b', clean_title, re.I),
        'quantity': re.search(r'(\d+\s?(pcs|pack|set))', clean_title, re.I),
        'size': re.search(r'(\d+-\d+Y|\d+Y|size \d+)', clean_title, re.I),
        'demographics': [],
        'clothing_terms': []
    }

    # Step 4: Process main product name from first segment
    main_segment = segments[0].strip()
    main_words = []

    # Identify and preserve demographics/clothing terms
    for word in re.findall(r"[\w']+", main_segment):
        lower_word = word.lower()
        if lower_word in DEMOGRAPHICS:
            components['demographics'].append(word)
        elif lower_word in CLOTHING_TERMS:
            components['clothing_terms'].append(word)
        else:
            main_words.append(word)

    # Step 5: Build short title structure
    short_title = []

    # Add color first if present
    if components['color']:
        short_title.append(components['color'].group(0).title())

    # Add demographic info
    short_title.extend(components['demographics'])

    # Add main product words (limit to 4 words)
    if main_words:
        # Keep first 4 words maximum, prioritizing end of phrase
        if len(main_words) > 4:
            main_words = main_words[:2] + main_words[-2:]
        short_title.extend(main_words[:4])

    # Add clothing terms if missing
    if not components['clothing_terms']:
        for term in CLOTHING_TERMS:
            if term in clean_title.lower():
                components['clothing_terms'].append(term.title())

    short_title.extend(components['clothing_terms'])

    # Step 6: Add technical specs
    specs = []
    if components['quantity']:
        specs.append(components['quantity'].group(0).upper())
    if components['size']:
        specs.append(components['size'].group(0))

    if specs:
        short_title.append('- ' + ' '.join(specs))

    # Step 7: Final cleaning and length adjustment
    final_title = []
    seen_words = set()

    # Deduplicate while preserving order
    for word in short_title:
        lower_word = word.lower()
        if lower_word not in seen_words:
            seen_words.add(lower_word)
            final_title.append(word)

    # Join and truncate to 50 characters
    final_title = ' '.join(final_title)
    if len(final_title) > 50:
        # Preserve the last part with specs
        if '-' in final_title:
            parts = final_title.split('-')
            core = ' '.join(parts[0].split()[-2:])
            final_title = f"{core} -{parts[1]}"
        final_title = final_title[:50].rsplit(' ', 1)[0]

    return final_title.strip()

cleaned_df['short_title'] = cleaned_df['title'].apply(generate_short_title)
cleaned_df
```

```python
cleaned_df.loc[:, 'product_type_id'] = pd.to_numeric(cleaned_df['product_type_id'], errors='coerce').astype('Int64')
cleaned_df.loc[:, 'product_length'] = cleaned_df['product_length'].round(2)
```

```python
# prompt: save each visualization as a png file

# Visualizations
# 1. Missing Values Comparison
plt.figure(figsize=(10, 6))
df.isnull().sum().plot(kind='bar', label='Before Cleaning', color='skyblue')
cleaned_df.isnull().sum().plot(kind='bar', label='After Cleaning', color='orange', alpha=0.7)
plt.title('Missing Values Before and After Cleaning')
plt.xlabel('Columns')
plt.ylabel('Number of Missing Values')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.savefig('missing_values_comparison.png')
plt.close()

# 2. Duplicate PRODUCT IDs
plt.figure(figsize=(8, 5))
plt.bar(['Before Cleaning', 'After Cleaning'], [df.duplicated(subset='PRODUCTID').sum(), cleaned_df.duplicated(subset='product_id').sum()],
        color=['skyblue', 'lightcoral'])
plt.title('Impact of Duplicate Removal on PRODUCT IDs')
plt.ylabel('Number of Duplicates')
plt.savefig('duplicate_product_ids.png')
plt.close()

# 3. Title Length Distribution
plt.figure(figsize=(10, 6))
sns.histplot(cleaned_df['title'].str.len(), kde=True, label='Original Titles', color='skyblue')
sns.histplot(cleaned_df['short_title'].str.len(), kde=True, label='Short Titles', color='orange', alpha=0.7)
plt.title('Title Length Distribution')
plt.xlabel('Title Length')
plt.ylabel('Frequency')
plt.legend()
plt.savefig('title_length_distribution.png')
plt.close()

# 4. Negative Value Counts
plt.figure(figsize=(10, 6))
(df.select_dtypes(include=['number']) < 0).sum().plot(kind='bar', label='Before Cleaning', color='skyblue')
(cleaned_df.select_dtypes(include=['number']) < 0).sum().plot(kind='bar', label='After Cleaning', color='orange', alpha=0.7)
plt.title('Negative Values Before and After Cleaning')
plt.xlabel('Numeric Columns')
plt.ylabel('Number of Negative Values')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.savefig('negative_value_counts.png')
plt.close()

print("Visualizations saved as PNG files.")
```
